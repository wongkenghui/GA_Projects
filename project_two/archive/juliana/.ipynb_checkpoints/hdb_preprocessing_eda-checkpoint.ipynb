# 3. Data Cleaning and Pre-Processing

## 3.1 Data Cleaning: Missing Values

### 3.1.1 Removing ID column and Combine Train & Test for Data Cleaning

```python
#Check the numbers of samples and features
print("The train data size before dropping ID feature is : {} ".format(df_train.shape))
print("The test data size before dropping ID feature is : {} ".format(df_test.shape))

#Save the 'id' column
train_ID = df_train['id']
test_ID = df_test['id']

#Now drop the  'id' colum since it's unnecessary for  the prediction process.
df_train.drop("id", axis = 1, inplace = True)
df_test.drop("id", axis = 1, inplace = True)

#Check again the data size after dropping the 'id' variable
print("\nThe train data size after dropping ID feature is : {} ".format(df_train.shape)) 
print("The test data size after dropping ID feature is : {} ".format(df_test.shape))
```

```python
ntrain = df_train.shape[0] 
ntest = df_test.shape[0] 
y_train = df_train['resale_price'] #Keep the target feature of train dataset
all_data = pd.concat((df_train, df_test)).reset_index(drop=True)
all_data.drop(['resale_price'], axis=1, inplace=True)
print("all_data size is : {}".format(all_data.shape))
```

### 3.1.2 Missing Values Handling

```python
#Check null values
all_data.isnull().sum().sort_values(ascending=False)[:15]
```

```python
#Check null values %
all_data.isnull().sum().sort_values(ascending=False)[:15]/len(all_data)*100
```

**Noted that 'Hawker_Within_500m', 'Mall_Within_500m', 'Hawker_Within_1km', 'Hawker_Within_2km', 'Mall_Within_1km' and 'Mall_Within_2km' actually mean there is no hawker or mall within the certain distance, so it can be filled with 0.** 

```python
#Fill the hawker/mall number with 0
all_data['Hawker_Within_500m'] = all_data['Hawker_Within_500m'].fillna(0)
all_data['Mall_Within_500m'] = all_data['Mall_Within_500m'].fillna(0)
all_data['Hawker_Within_1km'] = all_data['Hawker_Within_1km'].fillna(0)
all_data['Hawker_Within_2km'] = all_data['Hawker_Within_2km'].fillna(0)
all_data['Mall_Within_1km'] = all_data['Mall_Within_1km'].fillna(0)
all_data['Mall_Within_2km'] = all_data['Mall_Within_2km'].fillna(0)
```

```python
#Check null values again
all_data.isnull().sum().sort_values(ascending=False)[:15]
```

```python
all_data['Mall_Nearest_Distance'].isnull().sum()/len(all_data['Mall_Nearest_Distance'])*100
```

**The remain missing values in 'Mall_Nearest_Distance' mean there is no mall within 2km near the flat (i.e. the nearest mall is >=2km). These missing values can be filled with the possible min distance of the nearest mall 2000 (m). And these values are lesser than 1% of the total values so it will not impact the overall results.** 

```python
#Fill the NAN in 'Mall_Nearest_Distance' column with 2000
all_data['Mall_Nearest_Distance'] = all_data['Mall_Nearest_Distance'].fillna(2000)
```

## 3.3 Standardize Column Names and Save Final Dataframes

```python
#Get the column names and store as a list
column_names = list(all_data_selected.columns)
```

```python
#Downcase the column names
column_names_update = [x.lower() for x in column_names]
```

```python
#Replace the space with underscore in column names
column_names_update = [x.replace(' ', '_') for x in column_names_update]
```

```python
#Rename the column names in dataframe
column_renames = {key: value for key, value in zip(column_names, column_names_update)}
all_data_selected = all_data_selected.rename(columns=column_renames)
```

# 4. EDA

```python
plt.figure(figsize=(14, 6))
sns.set(font_scale=1)
sns.histplot(y_train,kde=False,bins=50)  

plt.title("HDB Resale Price Distribution")
plt.xlabel("Resale Price in SGD");
```

***Interpretation: The above image shows that most resale prices are in a range of 360,000 - 440,000 SGD. A strong right skewed means some resale prices are relatively high.*** 

```python
corr = round(df_train.corr(numeric_only=True),2)
k = 10

plt.figure(figsize=(40,30))
cols_10 = corr.nlargest(k, 'resale_price')['resale_price'].index
corrs_10 = df_train[cols_10].corr()
mask = np.zeros_like(corrs_10)
mask[np.triu_indices_from(mask)] = True
sns.set(font_scale=2)
# Plot our correlation heatmap, while masking the upper triangle to be white.
with sns.axes_style('white'):
    sns.heatmap(corrs_10, mask = mask, square = True, annot = True, vmin = -1, vmax = 1, linewidths = .5)
plt.title("Heatmap for Top 10 Correlated Features Selected")
```

***Interpretation: The heatmap above shows the top 5 features correlated to resale price are:***
1. floor_area_sqft
2. max_floor_lvl
3. 5room_sold
4. mid_storey
5. year_completed

```python
flat_type_list = ['1 ROOM','2 ROOM','3 ROOM','4 ROOM','5 ROOM', 'EXECUTIVE', 'MULTI-GENERATION']
```

```python
df_train_dict = {}
for flat_type in flat_type_list:
    df_train_dict[flat_type] = df_train[df_train['flat_type'] == flat_type]
```

```python
def plot_heatmap(key, df_train):

    corr = round(df_train.corr(numeric_only=True),2)
    k = 5

    plt.figure(figsize=(8,6))
    cols_5 = corr.nlargest(k, 'resale_price')['resale_price'].index
    corrs_5 = df_train[cols_5].corr()
    mask = np.zeros_like(corrs_5)
    mask[np.triu_indices_from(mask)] = True
    sns.set(font_scale=1)
    # Plot our correlation heatmap, while masking the upper triangle to be white.
    with sns.axes_style('white'):
        sns.heatmap(corrs_5, mask = mask, square = True, annot = True, vmin = -1, vmax = 1, linewidths = .5)
    plt.title(f"Heatmap for Top 5 Correlated Features of {key}")
```

```python
for key, df_train in df_train_dict.items():
    plot_heatmap(key, df_train)
```

***Interpretation: The heatmap above shows the top 5 features correlated to resale price for each flat type. Compared to the overall heatmap with all flat types included, the top 5 features correlated for each flat type are quite different. For example, the most correlated feature for each flat type is actually not the 'floor area in sqft', it is 'max floor level' for 3 Room and 4 Room and 'hawker numbers within 2km' for 5 Room, Executive and Multi-generation.***
